```python
search_times_with_partitioning = [3.3456, 2.3419, 2.2750]
search_times_without_partitioning = [3.3456, 1.7535, 1.4604]
num_servers = [1, 2, 3]

# Calculate speedup relative to single server
speedup_with_partitioning = [search_times_with_partitioning[0]/t for t in search_times_with_partitioning]
speedup_without_partitioning = [search_times_without_partitioning[0]/t for t in search_times_without_partitioning]

plt.plot(num_servers, speedup_with_partitioning, label='With Partitioning')
plt.plot(num_servers, speedup_without_partitioning, label='Without Partitioning')
# plot ideal speedup, gray dashed line
plt.plot(num_servers, [1, 2, 3], label="Ideal", color='gray', linestyle='--')
plt.xlabel('Number of Servers')
plt.xticks(num_servers)
plt.ylabel('Speedup')
plt.title('Strong Scaling Analysis on ARM')
plt.legend()
# make this as high quality as possible
plt.savefig('speedup.png', dpi=300)
plt.show()

```
